```python
# task 5
# Program to enter strings and display them in alphabetical order in lowercase

# Initialize an empty list to store strings
strings = []

# Prompt the user to enter strings until an empty input is given
while True:
    user_input = input("Enter a string (or press Enter to finish): ")
    if user_input == "":
        break
    strings.append(user_input)

# Convert all strings to lowercase and sort them alphabetically
strings = [s.lower() for s in strings]
strings.sort()

# Display the sorted list
print("\nStrings in alphabetical order (lowercase):")
for string in strings:
    print(string)
```
